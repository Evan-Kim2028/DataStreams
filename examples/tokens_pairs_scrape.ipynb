```python
from datastreams.datastream import Streamer

import matplotlib.pyplot as plt
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)
pd.set_option('max_colwidth', None)
```

```python
endpoint1 = 'https://api.thegraph.com/subgraphs/name/messari/sushiswap-ethereum' # sushi
endpoint2 = 'https://api.thegraph.com/subgraphs/name/messari/balancer-v2-ethereum' # balancer
```

```python
# load streamer class
ds1 = Streamer(endpoint1)
ds2 = Streamer(endpoint2)
```

```python
# create a list of Streamer objects
ds_list = [ds1, ds2]
```

```python
# check available query fields for a SINGLE endpoint
ds_list[0].queryDict.values()
```

```python
# check which QueryDict values are the same
ds_list[0].queryDict.keys() & ds_list[-1].queryDict.keys()
```

```python
tokens_list = []
pool_list = []
```

```python
#run token query on streamer objects
for ds in ds_list:
    # token query
    tokens_df = ds.runQuery(ds.queryDict['tokens'], query_size=123)
    tokens_list.append(tokens_df)
    # pool query
    liquidity_df = ds.runQuery(ds.queryDict['liquidityPools'], query_size=123)
    pool_list.append(liquidity_df)
```
