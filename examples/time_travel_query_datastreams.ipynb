```python
# select subgraph endpoint
endpoint = 'https://api.thegraph.com/subgraphs/name/messari/uniswap-v3-ethereum'
```

```python
# https://info.uniswap.org/#/pools/0x88e6a0c2ddd26feeb64f039a2c41296fcb3f5640
# https://etherscan.io/address/0x88e6a0c2ddd26feeb64f039a2c41296fcb3f5640

# USDC/WETH 
lp_name = '0x88e6A0c2dDD26FEEb64F039a2c41296FcB3f5640'
```

```python
# load Streamer class
ds = Streamer(endpoint)
```

```python
# check available query fields via the query dictionary.
ds.queryDict.keys()
```

```python
df = ds.runQuery(ds.queryDict.get('liquidityPools'), query_size=1500)
```
